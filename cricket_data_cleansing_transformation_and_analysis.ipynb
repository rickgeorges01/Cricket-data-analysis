```python
# Importe les bibliothèques nécessaires
import pandas as pd
import numpy as np
!pip install matplotlib
import matplotlib.pyplot as plt
```

```python
# Lecture du fichier Excel dans un DataFrame
df = pd.read_excel('CricketData.xlsx')

# Affiche les 20 premières lignes du DataFrame pour un aperçu initial
df.head(20)
```

```python
# Renomme certaines colonnes pour les rendre plus explicites
df = df.rename(columns={
    'Mat': 'Matches',
    'NO': 'Not Outs',
    'HS': 'Highest_Inns_Score',
    'BF': 'Balls_faced',
    'SR': 'Batting_Strike_Rate'
})

# Affiche à nouveau les 20 premières lignes pour vérifier le renommage
df.head(20)
```

```python
# Vérifie la présence de valeurs manquantes dans le DataFrame
missing_values = df.isnull().any()
print(missing_values)

# Affiche les lignes où la colonne 'Balls_faced' contient des valeurs manquantes
missing_balls_faced = df[df['Balls_faced'].isna()]
missing_balls_faced
```

```python
# Remplace les valeurs manquantes dans 'Balls_faced' par 0
df['Balls_faced'] = df['Balls_faced'].fillna(0)

# Vérifie si les valeurs manquantes ont été remplacées correctement
df[df['Player'] == 'CL Walcott (WI)']
```

```python
# Vérifie les doublons dans la colonne 'Player'
# autre façon : df.duplicated()
# autre façon : df[df['Player'].duplicated() == 1]
duplicated_players = df[df['Player'].duplicated()]
duplicated_players
```

```python
# Filtre pour des joueurs spécifiques afin de vérifier les doublons
players_to_check = ['AC Voges (AUS)', 'Mohammad Yousuf (PAK)', 'DCS Compton (ENG)', 'TT Samaraweera (SL)']
df[df['Player'].isin(players_to_check)]
```

```python
# Supprime les doublons
df = df.drop_duplicates()
# Vérifie à nouveau pour ces joueurs après suppression des doublons
df[df['Player'].isin(players_to_check)]
```

```python
# Sépare la colonne 'Span' en deux colonnes distinctes pour l'année de début et de fin de carrière
df['Rookie_Year'] = df['Span'].str.split('-').str[0].astype(int)
df['Final_Year'] = df['Span'].str.split('-').str[1].astype(int)
# Supprime la colonne d'origine 'Span'
del df['Span'] 
```

```python
# Affiche les 10 premières lignes après ces modifications
df.head(10)
```

```python
# Extrait le pays du joueur de la colonne 'Player'
df['Country'] = df['Player'].str.split('(').str[1].str.split(')').str[0]

# Supprime le pays du nom du joueur pour ne garder que le nom
df['Player'] = df['Player'].str.split('(').str[0].str.strip()
```

```python
# Affiche les 10 premières lignes après ces modifications
df.head(10)
```

```python
# Changer le type de certaines colonnes
df = df.astype({'Rookie_Year': 'int', 'Final_Year': 'int'})

#df['Highest_Inns_Score'].str.split(pat = '*').str[0]
#df['Highest_Inns_Score'] = df['Highest_Inns_Score'].str.split(pat = '*').str[0]
#df['Highest_Inns_Score'] = df['Highest_Inns_Score'].astype('int')
#df['Highest_Inns_Score']

# Extraire et convertir 'Highest_Inns_Score' en entier, en ignorant les astérisques
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].str.split('*').str[0].astype(int)


# Extraire et convertir 'Balls_faced' en entier, en ignorant les caractères après un "+"
df['Balls_faced'] = df['Balls_faced'].astype(str).str.split('+').str[0].astype(int)

# Vérifier les types de données après conversion
df.dtypes

# Calculer la durée de la carrière de chaque joueur en années
df['career_length'] = df['Final_Year'] - df['Rookie_Year']

# Afficher le DataFrame complet pour vérifier toutes les modifications
df.head(10)
```

```python
# Remplacer les valeurs spécifiques dans la colonne 'Country'
df['Country'] = df['Country'].replace({
    'AUS': 'Australia',
    'ENG': 'England',
    'IND': 'India',
    'PAK': 'Pakistan',
    'NZ': 'New Zealand',
    'WI': 'West Indies',
    'SA': 'South Africa',
    'SL': 'Sri Lanka',
    'BAN': 'Bangladesh',
    'ZIM': 'Zimbabwe',
    'AFG': 'Afghanistan',
    'IRE': 'Ireland',
    'ICC/PAK': 'Pakistan',
    'ICC/SA': 'South Africa',
    'ICC/WI': 'West Indies',
    'ICC/IND': 'India'
})

# Vérifier les remplacements
df['Country'].unique()
```

```python
# Ajouter des colonnes dérivées basées sur les performances
df['Matches_Per_Year'] = df['Matches'] / df['career_length']
df['Runs_Per_Match'] = df['Balls_faced'] / df['Matches']

# Afficher les nouvelles colonnes pour vérifier les transformations
df[['Player', 'Matches_Per_Year', 'Runs_Per_Match']].head(10)
```

```python
# Créer des intervalles pour les périodes de carrière
bins = [0, 1940, 1960, 1980, 2000, 2020]
labels = ['Avant 1940', '1940-1960', '1960-1980', '1980-2000', 'Depuis 2000']
df['Career_Era'] = pd.cut(df['Rookie_Year'], bins=bins, labels=labels)

# Afficher les résultats du binning
df[['Player', 'Rookie_Year', 'Career_Era']].head(10)
```

```python
# Filtrer les joueurs qui ont joué plus de 50 matchs et ont un taux de frappe supérieur à 50
filtered_df = df[(df['Matches'] > 50) & (df['Batting_Strike_Rate'] > 50)]

# Afficher les joueurs filtrés
filtered_df[['Player', 'Matches', 'Batting_Strike_Rate']].head(10)
```

```python
# Calcule et affiche la durée moyenne de la carrière
average_career_length = df['career_length'].mean()
print(f'Durée moyenne de la carrière : {average_career_length} ans')

# Calcule et affiche la moyenne du taux de frappe pour les joueurs ayant une carrière de plus de 10 ans
long_career_strike_rate = df[df['career_length'] > 10]['Batting_Strike_Rate'].mean()
print(f'Moyenne du taux de frappe pour les carrières de plus de 10 ans : {long_career_strike_rate}')
```

```python
# Analyse des statistiques moyennes des joueurs par pays
country_stats = df.groupby('Country').agg(
    Average_Matches=('Matches', 'mean'),
    Average_Batting_Strike_Rate=('Batting_Strike_Rate', 'mean'),
    Average_Career_Length=('career_length', 'mean')
).reset_index()

country_stats
```

```python
# Trouver les 5 joueurs avec le plus grand nombre de matchs
top_players_matches = df.sort_values(by='Matches', ascending=False).head(5)
print("Joueurs avec le plus grand nombre de matchs :")
top_players_matches[['Player', 'Matches', 'Country']]
```

```python
# Trouver les 5 joueurs avec la plus longue carrière
top_players_career_length = df.sort_values(by='career_length', ascending=False).head(5)
print("Joueurs avec la plus longue carrière :")
top_players_career_length[['Player', 'career_length', 'Country']]
```

```python
# Trouver les 5 meilleurs frappeurs (Strike Rate) en excluant les valeurs manquantes
top_batters_strike_rate = df[df['Batting_Strike_Rate'].notna()].sort_values(by='Batting_Strike_Rate', ascending=False).head(5)
print("Meilleurs frappeurs selon le taux de frappe :")
top_batters_strike_rate[['Player', 'Batting_Strike_Rate', 'Country']]
```

```python
# Comparer le taux de frappe des joueurs ayant une carrière courte (<5 ans) et longue (>=5 ans)
short_career_strike_rate = df[df['career_length'] < 5]['Batting_Strike_Rate'].mean()
long_career_strike_rate = df[df['career_length'] >= 5]['Batting_Strike_Rate'].mean()

print(f"Moyenne du taux de frappe pour les joueurs avec une carrière courte (<5 ans) : {short_career_strike_rate}")
print(f"Moyenne du taux de frappe pour les joueurs avec une carrière longue (>=5 ans) : {long_career_strike_rate}")
```

```python
# Trouver les 5 joueurs avec le plus grand nombre de centuries (scores de 100 ou plus)
top_century_players = df[df['Highest_Inns_Score'] >= 100].sort_values(by='Highest_Inns_Score', ascending=False).head(5)
print("Joueurs avec le plus grand nombre de centuries (scores de 100 ou plus) :")
top_century_players[['Player', 'Highest_Inns_Score', 'Country']]
```

```python
# Analyser le ratio de scores par 'Matches' et 'Balls_faced'
df['Runs_per_Match'] = df['Balls_faced'] / df['Matches']
df['Runs_per_Ball'] = df['Balls_faced'] / df['Balls_faced']

# Calculer la moyenne de ces ratios
average_runs_per_match = df['Runs_per_Match'].mean()
average_runs_per_ball = df['Runs_per_Ball'].mean()

print(f"Moyenne des runs par match : {average_runs_per_match}")
print(f"Moyenne des runs par balle : {average_runs_per_ball}")
```

```python
# Filtrer les joueurs actifs au cours des 5 dernières années
recent_players = df[df['Final_Year'] >= 2019]

# Trouver les meilleurs joueurs en fonction du taux de frappe
top_recent_players = recent_players.sort_values(by='Batting_Strike_Rate', ascending=False).head(5)
print("Meilleurs joueurs sur les 5 dernières années (par taux de frappe) :")
top_recent_players[['Player', 'Batting_Strike_Rate', 'Country', 'Final_Year']]
```

```python
# Calculer la performance moyenne de chaque pays en termes de Matches et Batting Strike Rate
team_performance = df.groupby('Country').agg(
    Average_Matches=('Matches', 'mean'),
    Average_Batting_Strike_Rate=('Batting_Strike_Rate', 'mean')
).sort_values(by='Average_Batting_Strike_Rate', ascending=False)

print("Comparaison des performances entre les équipes :")
team_performance
```
